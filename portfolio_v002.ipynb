```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tickers = ["BABA", "3690.HK", "0700.HK", "1810.HK", "1658.HK"]
data = yf.download(tickers, start="2022-01-01", end="2023-05-26", group_by='ticker')
close_data = pd.concat([data[ticker]['Close'] for ticker in tickers], axis=1, keys=tickers)

returns_data = close_data.pct_change()
corr_matrix = returns_data.corr()

sns.clustermap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Selected Stocks")
plt.show()
```
